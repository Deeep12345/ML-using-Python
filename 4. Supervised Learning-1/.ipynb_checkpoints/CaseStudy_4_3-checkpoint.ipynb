```python
import pandas as pd 
from sklearn.model_selection import train_test_split # to split the data into two parts
from sklearn import metrics # for the check the error and accuracy of the model
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import random
import matplotlib.pyplot as plt # this is used for the plot the graph
import seaborn as sns

# read data
data = pd.read_csv('loan_borowwer_data.csv')

# convert categorical data into numeric
data.purpose = pd.Categorical(data.purpose)
data['purpose'] = data.purpose.cat.codes

# outcome count and orrelation plot 
sns.countplot(data['not.fully.paid'],label="Count")
plt.show()
corr = data.corr() # .corr is used for find corelation
plt.figure(figsize=(14,14))
sns.heatmap(corr, cbar = True,  square = True,
            cmap= 'coolwarm')
plt.show()

# drop columns which will be not part of the training/test model
random.seed(2)
pred_columns = data[:]
pred_columns.drop(['not.fully.paid'],axis=1,inplace=True)
#pred_columns.drop(['purpose'],axis=1,inplace=True)
#pred_columns.drop(['int.rate'],axis=1,inplace=True)
#pred_columns.drop(['installment'],axis=1,inplace=True)
prediction_var = pred_columns.columns

# create train/test data
train, test = train_test_split(data, test_size = 0.3)# in this our main data is splitted into train and test
train_X = train[prediction_var]# taking the training data input
train_y= train['not.fully.paid']# This is output of our training data

# same we have to do for test
test_X= test[prediction_var] # taking test data inputs
test_y =test['not.fully.paid']   #output value of test dat

# RandomForest classifier
model=RandomForestClassifier(n_estimators=100)# a simple random forest model
model.fit(train_X,train_y)# now fit our model for traiing data
prediction=model.predict(test_X)# predict for the test data
print('RandomForest accuracy :' )
print(metrics.accuracy_score(prediction,test_y)) # to check the accuracy

# Decision Tree 
model = tree.DecisionTreeClassifier()
model.fit(train_X,train_y)
prediction=model.predict(test_X)
print('DecisionTree  accuracy : ' )
print(metrics.accuracy_score(prediction,test_y))
```
